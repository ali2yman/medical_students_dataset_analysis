```python
df = df.dropna()
df.head()
```

```python
null_values = df.isnull().any()
null_values
```

```python
start_index = df[df['Student ID'] >= 215000].index.min()

# Check if there is a student with the specified id
if start_index is not None:
    # Update the ids starting from 215001
    df.loc[start_index:, 'Student ID'] += 1
else:
    print("No student found with id starting from 215000")
```

```python
df
```

```python
smoking_students = df[df['Smoking'] == 'Yes']

if not smoking_students.empty:
    df.loc[smoking_students.index, 'Smoking'] = 'Yes, Their lungs will damage'
    print("Updated smoking states for smoking students.")
else:
    print("No students found with smoking state 'Yes'.")

print(df['Smoking'])
```

```python
df["Gender"].value_counts()
```

```python
age_range = [10, 20, 30, 40, float('inf')]
age_labels = ['10-20', '21-30', '31-40', '41+']

age_counts = pd.cut(df['Age'], bins=age_range, labels=age_labels, right=True).value_counts().sort_index()

print("Number of students in each age range:")
print(age_counts)
```

```python
df['Height'] = df['Height'].round()
df['Height']
```

```python
df['Weight'] = df['Weight'].round()
df['Weight']
```

```python
df['Blood Type'].value_counts()
```

```python
d = {"Yes": 'student that has a Diabetes disease', "No":"HAS A GOOG HEALTH AND MUST BR CURD"}

df["Diabetes"] = df["Diabetes"].map(d)


print(df['Diabetes'])
```

```python
df
```

```python
def suitable_Fun(row):
    if abs((row['Height'] - 100) - row['Weight']) == 10:
        return "Suitable"
    else:
        return "Not Suitable"

df['suitable_body'] = df.apply(suitable_Fun, axis=1)
```

```python
df['suitable_body'] 
```

```python
label_encoder = LabelEncoder()

df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Blood Type'] = label_encoder.fit_transform(df['Blood Type'])
df['Diabetes'] = label_encoder.fit_transform(df['Diabetes'])
df['Smoking'] = label_encoder.fit_transform(df['Smoking'])
df['suitable_body'] = label_encoder.fit_transform(df['suitable_body'])
```
